```python
import numpy as np
dimension = int(input("Digita tamaño de la letra: "))
matrix = np.zeros((dimension,5))


digito = ""
for x in range(0, dimension):
  for y in range(0, dimension):
    
    if( y == x or y == (dimension - x - 1)):
      digito = digito + "X"
    else:
      digito = digito + "_"
  digito = digito + "\n"

print(digito)

```

```python
def calculateBlock(mylist):
  auxList = []
  result = ""
  arrResult = []
  lastElement = mylist[-1]
  contador = 0
  for element in mylist:  
    contador = contador + 1  
    if element != 0:
      auxList.append(element)
      if(contador == len(mylist)) and element == lastElement:
        auxList.sort()
        sort_str = "".join(map(str,auxList))
        result = result + sort_str + "  "
        auxList.clear()

    elif element == 0 and len(auxList) != 0:
      auxList.sort()
      sort_str = "".join(map(str,auxList))
      result = result + sort_str + "  "
      auxList.clear()   

    elif element == 0 and len(auxList) == 0:
      result = result + "X" + " "
      
  print(result)

```

```python
mylist =  [2,1,0,0,3,4 ]
calculateBlock(mylist)
```
